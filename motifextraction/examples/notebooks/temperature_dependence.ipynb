```python
# Calculate/load the VIs for each cluster for each temperature
vp_indices = {}
for T in temperatures:
    print(f"Temperature: {T}")
    if not Path((data_path / "cluster_vps.npy").format(T=T)).exists():
        vp_path = (data_path / "cluster_vps.npy").format(T=T)

        indices = np.zeros((NCLUSTERS, 10), dtype=np.int)
        indices.fill(np.nan)

        for i in range(NCLUSTERS):
            fn = (cluster_path / f'{i}.xyz').format(T=T)
            c = Cluster(filename=fn)
            indices[i, :] = c.vp_index
            if i % 1000 == 0:
                print(i); sys.stdout.flush()
        np.save(str(vp_path), indices)
    else:
        indices = np.load(str(Path(data_path / "cluster_vps.npy").format(T=T)))
    vp_indices[T] = indices
```

```python
# Load the motif errors
def get_motif_errors(motif, T):
    stem = motif.filename.stem
    error_path = Path(motif.data_path.format(T=T)) / f"motif_errors/{stem}_errors.npy"
    return np.load(error_path)

with open("../motifextraction/examples/motifextraction/metallic_glass/data/norm_factors.json") as f:
    nf = json.load(f)
for motif in sorted(motifs, key=lambda m: m.CN):
    print(motif.CN, motif.filename.stem, motif.vp_index[:4])
    motif.errors = {}
    for T in temperatures:
        errors = get_motif_errors(motif, T)
        # Run the norm-factor rescaling on the errors
        L2 = errors[:, 0]
        L1 = errors[:, 1]
        Linf = errors[:, 2]
        angular = errors[:, 3]
        L2 /= nf['L2']['divide_by']
        L1 /= nf['L1']['divide_by']
        Linf /= nf['Linf']['divide_by']
        angular /= nf['angular']['divide_by']
        g = scipy.stats.mstats.gmean([L2, L1, Linf, angular])
        g[np.where(np.isinf(g))] = np.nan
        motif.errors[T] = g
```

```python
motif_errors = np.zeros((len(motifs), len(temperatures), NCLUSTERS))
motif_errors.fill(np.nan)
for m, motif in enumerate(motifs):
    for t, T in enumerate(temperatures):
        motif_errors[m, t, :] = motif.errors[T]
print(motif_errors.shape)
```

```python
# Assign
def assign(c, motifs, T):
    e = np.inf
    idx = np.nan
    for i, motif in enumerate(motifs):
        if np.abs(cluster_cns[T][c] - motif.CN) > 0:  # Enforce CN similarity +- 0
            continue
        if motif.errors[T][c] < e:
            e = motif.errors[T][c]
            idx = i
    return idx, e

assignments = np.zeros((len(temperatures), NCLUSTERS))
assignments.fill(np.nan)
for t, T in enumerate(temperatures):
    print(T)
    for c in range(NCLUSTERS):
        assignments[t, c] = assign(c, motifs, T)[0]
```
